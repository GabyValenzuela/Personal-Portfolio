<font color=black> **Welcome to the SpaceX-213 Game** </font>


<span><small>The purpose of this game is to take the user into a spaceship trip, which has the objective of reaching planet N20-XF809 that is supposed to be inhabitated by another intelligent life. The challenge of the game will be to make the user reach that planet or if that is not possible, to connect with this intelligent life and come back home with all its crew. During this trip the user will be given a couple of choices that will take them to a win or failure outcome.</small></span>

</span><small>This game is intended to be fun for the user and to explore different scenarios while playing it a second or third time.</small></span> 

</span><small>This game contains some truthful aspects such as the names of the planets, and some others based on the imagination of the creator.</small></span>

</span><small>Hope you enjoy the game!</small></span>


```python

################################################################################
################################################################################
######################## WELCOME TO SPACESHIP GAME #############################
################################################################################
############This is a spaceship mission that I hope you enjoy!!!!###############
################################################################################
################################################################################

#This game has 5 main stages:
#         Mission Space
#         Traveling Space
#         Out of Via Lactea
#         Planets
#         Uncertainty
#As well as the winning and failing scenarios
#
#In the very first two stages, the user will keep advancing to other stages\
#and only as they are "Out of Via Lactea" they will take their very fist decision.
#
#In the meantime, the user will get messages about the mission and also a counting\
#about the launching.
#
#I really hope any user enjoys this journey that I prepared.
#Whitouth further ado... LET'S START THIS GAME!!!!

###################################################################################
###################################################################################

### Inserting a complement to slow down the first part of the code ###

import time
import os, sys, time

### Slowing time for text
time.sleep(1)

### UDF for slowing timw
def print_slow(s, tail):
   for letter in s:
      print(letter, end="")
      sys.stdout.flush()
      time.sleep(.03)
   print(end=tail)

### Welcoming the user and giving it a hint of what the game will be about
print_slow(f"""Welcome to SpaceX-213 adventure!\n
In this game you'll prove your ability as a commander of a spaceship.\n""","")
time.sleep(1)

print_slow(f"""\nYour mission if you accept it, is to reach planet N20-XF809 that is supposed to be
inhabitated by another intelligent life and to make contact with them.

Either way if you reach the planet or make contact with intelligent life your mission
will be completed.\n""","")

### Let's build some emotion to the user
print_slow(f"""\nAre you ready to play?\n""","")
time.sleep(1)

### Making the user interact for the first time with an enter ###
print(f"""\n< Please type enter to start >""")
input(prompt = f""" """) ############################################################################# This is where I entered the extra space \n
time.sleep(1)

### Slowed random message to the game, so that at the beginning this builds emotion ###

def mayday_message(str):
   for letter in str:
       sys.stdout.write(letter)
       sys.stdout.flush()
       time.sleep(.01)

### Printing a message to the user that will keep it excited
mayday_message(f"""....Mayday mayday we need assistance here!!!...\n""")

### Slowing other messages so that text does not appear instantly
print(f""" \n< Please type enter to continue >""")
input(prompt = f""" """)
time.sleep(1)

#################################################################################
################################ MISSION SPACE STAGE ############################

### Taking user to the first stage... THE MISSION!!!

###Defining a global variable so that the user can be named at the end. 
#Thus, personalizing the whole experience of the game
capitain_name_2 = ""

###Defining the mission space variable
def mission_space():
    #Again the global variable as it will be created here
    global capitain_name_2
    #Message of what the mission is about
    print_slow(f"""
    It's 2100 here on Earth,and you are the commander of Space X-213.\n
    Yes, same type as that one launched by Tesla in 2020, but after 212 more launches\n\
    and many, many more improvements!\n
    Your mission as a commander is to arrive to a new planet called N20-XF809 which is\n\
    in Hoag's Object Galaxy.\n ""","")

    #Giving the user a space for reading
    print("""\n<Please type enter to continue> """)
    input(prompt = "")
    
    #Main objective of the mission/game
    print_slow(f"""
    Your mission is to go there and investigate an intelligent life that sent a message
    two months ago and to come back safe and sound with all your crew.
    
    This trip won't be easy as a leader you'll have to make hard decisions that may
    or may not bring you back. ""","")
    
    #Personalizing the game
    print(f"""
    \nBut before we start, please Capitain, tell me your name:\n """)
    
    #Finally, defining the global variable which will enhace user expericience
    capitain_name = input(prompt = "")
    #Capitalizing the variable so that the name of the user is standarized 
    capitain_name_2 = capitain_name.capitalize()
    
    #Slowing the message of the launching mission
    def launching_time(str):
        for number in str:
           sys.stdout.write(number)
           sys.stdout.flush()
           time.sleep(.02)   
    
    #Launching message
    launching_time(f"""\n1\n\n2\n\n3...\n
    Launching Capitain {capitain_name_2}!!!\n """)
    
    #Giving the user a space for reading
    print(""" \n<Please type enter to continue>""")
    input()
    
    #Taking the user to the traveling space stage - which will be the first in which the user will be tested 
    traveling_space()
    
    #################################################################################
############################### TRAVELING SPACE STAGE ###########################

### Defining stage... TRAVELING SPACE!!!
def traveling_space():
    ### Letting know the user that they are in the TRAVELING SPACE stage!
    print_slow(f"""
    You're finally in space and there have been a couple of hours since you left Earth...
    
    How does it make you feel?""","")
    
    ### Testing user excitement
    print_slow(f"""\n
    1. Excited! Never been better before!\n
    2. Fearful! What if something goes wrong?\n
    3. Not so sure...
    
    < Please type your answer > ""","")
    
    ###Conditional answer in regards to its level of excitement
    feeling = input(prompt = "")
    
    ###Adding rules so that the user can type a number or a word, even if a number was requested
    if feeling == "1" or feeling == "1." or feeling == "excited" or "excited" in feeling or "one" in feeling:
        print_slow(f"""
        GREAT! That is the adventure spirit!\n""","")
    elif feeling == "2" or feeling == "2." or feeling == "fearful" or "fearful" in feeling or "two" in feeling:
        print (f"""
        Well, you're right to be worried, lots of things can and will go wrong!\n""")
    ###Last condition that opens up so that the user that wasn't excited gets excited
    elif feeling == "3" or feeling == "3." or feeling == "not so sure" or "not sure" in feeling or "three" in feeling:
        print (f"""
        That's understandable, this is a whole new experience... What if I tell you
        that you stand really good chances of completing this mission.
        
        Will that give you more hope?\n
        < Please type Y/N >\n""")
        
        ###User's input for insecure expectations
        motiviation_answer= input(prompt= "")
        
        ###Analizing user's answer - nested conditional
        ###This condition won't affect the path that the user will follow, it will only try to empathize
        if motiviation_answer == 'yes' or motiviation_answer == "Y" or "y" in motiviation_answer :
            print("""\n\nHooray!!! That's such a good attitude!!!\n""")
        else :
            print("""\n\nWell, hopefully you'll be more enthusiastic as you spend more time on space.\n""")
    
    ###If the user inputs sth different than what was asked, they will still continue with the game...
    ###...but the message will be different
    else :
        print (f"""\nI'm sorry, I didn't get how you feel""")
    ###Sending user to the next stage after a message according to its feelings was displayed
    out_of_vialactea()

#################################################################################
################################## OUT OF VIA LACTEA ############################

### Defining stage... OUT OF VIA LACTEA!!!   
def out_of_vialactea():
    #Printing first life or death decision  
    print_slow(f"""
    You've just passed the Via Lactea Galaxy belt, a crucial and difficult task\n\
    that you managed well except for the fact that your orbiter was hit by a meteorit\n\
    that led you to take one of the following actions...
    
    1. Ask for assistance and call home to go back\n
    2. Go to the closest planet and wait for Earth to send spare parts,which will let you \n\
    continue with your mission.\n
    3. Let know to the station that you'll continue until the spaceship can no longer go\n
    
    < Please type the number of your choice >\n""","")
    

    #Controlling input for life or death decision variable
    while True:
        dec_lof = input(prompt = "")
    #Controloing user's input in regards to this life or death decision
    #Only numbers allowed
    #Their decision will take them to a different stage
        if dec_lof.isnumeric():
            break
        else:
            print("Please insert a numerical value")
            
    #Their decision will take them to a different stage
    if dec_lof == "1" :
        print (f"""
        Well, that was the safest decision. You're now heading home
        \n""")
        #Taking user to failure stage
        fail()
    elif dec_lof == "2" :
        print (f"""There are four options of planets nearby:\n
        1. Dagon, known for being a really hot planet\n
        2. Tadmor, known for its frozen temperatures\n
        3. Kapteyn b, known as the oldest planet discovered in the Universe\n
        4. Kepler-62f, known for its deep craters on the surface\n""")
        #Taking user to planet stage
        planets ()
    elif dec_lof == "3":
        print (f"""You really took an intrepid approach!\n""")
        #Taking user to uncertainty stage
        uncertainty()
    #If user inputs an invalid number, this will take them back to this stage
    else :
        print("""Please type a number between 1 and 4""")
        out_of_vialactea()
    
    
##################################################################################
############################## PLANETS STAGE #####################################

#Defining PLANETS stage!!!!
def planets():
    #Asking the user for its answer from the previous stage
    print("""Which planet would you like to visit?\n
          < Please type the number of your option >""")
    
    #Defining the loop for controlling number inputs
    while True:
        #defining the variable
        selected_planet = input(prompt = "")
        #Controlling if its a number
        if selected_planet.isnumeric():
            #Break the loop when the input is a number
            break
        #If user does not insert a numerical value, then it will be asked again
        else:
            print("Please insert a numerical value")
    #The following loop will guide the user till the right answer
    #Only number 3 will move then to the next stage
    while selected_planet != "3" :
        #Print statements that suggest the user to try another number
        print("""It looks like that planet wasn't ideal. Why don't you give it another try\n""")
        print(""""< Please try with another number >""")
        #New check of the new input inserted
        while True:
            selected_planet = input(prompt = "")
            #This loop is the check for the following trials of the user, so that they enters a number
            if selected_planet.isnumeric():
            #If the value is a number, the user can continue with the game   
                break
            #If the user inserts their second time a non numerical value, a message will be released again
            else:    
                print("......Please insert a numerical value.......")
    
    #When the user finally enters the right choice of the game
    print("""That's right "Kapteyn b" is the most convinient planet. You're back\n\
    on track towards planet N20-XF809\n""")
    #Sending the user to the uncertainty stage
    uncertainty()
    
#################################################################################
########################## UNCERTAINTY STAGE #####################################

#Defining UNCERTAINTY stage!!! - Stage that will lead the user to a win or fail stage!!!
def uncertainty():
    #Message and question for uncertainty stage
    print(f"""\n
        Now, you realize that now you are heading towards a black hole...\n
        1. You could keep heading towards the black hole\n
        2. Head towards a spaceship that doesn't look like made by humans\n\n
        < Please type the number of your next action >\n""")
    #Input for user's decision
    while True:
        un_decision = input()
        #Controloing user's input in regards to their uncertain decision
        #Only numbers allowed
        #Their decision will take them to fail or win stage
        if un_decision.isnumeric():
            break
        else:
            print("Please insert a numerical value")

    #Analizing decision with some possible outcomes
    if un_decision == "1" :
            print("""Oh no, you spaguetized SpaceX-213 and all your crew\n""")
            #Sending the user to the failure stage
            fail()
    elif un_decision == "2" :
            print ("""How lucky!!! The spaceship was actually from N20-XF809 inhabitants\n""")
            #Sending the user to the winning stage
            win()
    #If user does not input a valid number, the program won't let it continue
    else :
            print(""" Please enter 1 or 2 """)
            #Sending user to the beggining of the stage until they insert the required input
            uncertainty()
        
#################################################################################
############################ WINNING STAGE - END OF THE GAME ####################

#Defining WINNING stage!!!
def win():
    #Making use of the global variable so that the message comes personalized
    global capitain_name_2

    #Final message - Winner message - if the user wins, they won't be asked to play again
    print(f"""
    CONGRATULATIONS Capitain {capitain_name_2}!!! You made it till the end!!!\n
    You were able to communicate with inhabitants of planet N20-XF809, which by the way, were really nice\n\
    and you even got back home with a selfie with them.\n\n
    WELL DONE!!!""")
    
#################################################################################    
############### FAILURE STAGE - PLAY AGAIN OR END OF THE GAME ###################

def fail():
    #Making use of the global variable so that the message comes personalized
    global capitain_name_2
    
    #Looser message
    print(f"""
    I'm afraid to inform you Capitain {capitain_name_2}, that you didn't complete the mission \n\
    and becuase of that reason the Earth was invaded by the inhabitants of planet N20-XF809.\n
                                        So baaaad!!!""")
    
    #Giving the user another try
    print(f"""
    Capitain {capitain_name_2}: Would you like to give it another try? (Y/N)\n
    """)
    play_again = input(prompt = "> ")
    play_again = play_again.lower()
    
    #If the user wants to play again, it will accept certain letters according\
    #to what was given in the instructions
    if play_again == 'yes' or play_again == "Y" or "y" in play_again :
        #if a user wants to play again, they will be sent directly to the first stage
        mission_space()
        
    #Final message - if the user decides not to play again or if they enter an invalid answer    
    else:
        print(f"""\n\nThanks for playing {capitain_name_2}...\n\n
        It was a pleasure to interact with you today!!!\n\n""")
        #exit(0)        
        
#################################################################################
#################################################################################

#Executing the whole mission function ******GAME*****       
mission_space()
```
