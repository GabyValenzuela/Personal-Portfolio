Northern Africa is known for some geographical aspects such as the Atlas Mountains, the connection to the Nile river as well as vast areas surrounded by desert. Culturally, it is dominated by a mix of Arabic and Sub-Saharan cultures with a strong family-centered way of living. Although this area has some economically stable countries such as Egypt and Morocco, some of these have been in conflicts for years like Libya (since 2011), Chad (2005-2010), Mali (2012-2015), and Sudan (civil war since 2013, just ended in February 2020). Some of the former events were caused by political reasons, which contributed to altering the economic stability of the countries and their development.     

The region as a whole has a very unique standpoint in a cultural aspect, as the region strongly depends on SMEs, which represent more than 77% of the jobs in Africa. The region has the potential to become a strong market in the upcoming decade due to the lack of supply of westernized services and thus is a target area for foreign investors.

Firstly, we defined obscure findings as anything that is not considered “normal” for the countries presented in our regions. To that end, our group categorized outliers, extreme values, and discrepancies in this category. To have a solid understanding of our dataset, we performed research via 3rd party sources before performing any analyses.
 
Secondly, to create a structure for the variables, we created three main categories in which we regrouped various factors: 
- Economy
- Social
- Health
 
In the Economy sector, we found some discrepancies in the data that included extreme values, which seemed unrealistic to be used for analysis. One of these cases was in the “GDP per employed person” column, where Libya had a value of $95,442 when the mean of the 10 countries was only $26,884. After analyzing the discrepancy, our team arrived at the assumption that this value might be misleading because employment numbers might be incorrectly reported and are lower in the calculation than they are in reality. Also, Chad had extreme values almost throughout all of the categories as it is very underdeveloped and also lacks proper data collection.
 
For the Social sector, Niger and Chad had irregular values. The literacy rate was reported at 28% and 29% of the population, respectively, as neither countries have the means and resources to provide adequate education to children. Besides, only about 1% of the population has access to the internet in these countries. We conducted a regression analysis to observe a strong correlation of 0.56 to exists between the two variables mentioned above, which is a sign of underdevelopment. 
 
Lastly, the Health sector also showed discrepancies in the reported cases of malaria. Some countries have data that equal to 0, which can be explained by the eradication of the disease in some areas. However, the problem is that some of these countries do not report external cases. If someone contracts malaria outside of the country itself, it is not counted as a positive case. To that end, the data given might be misleading.
 
Finally, we need to clarify that there are two other important events that may discredit the validity of some of the data reported. Libya’s data specifically has some information from the ‘90s as well as some data points that do not reflect its current situation. Moreover, before the 2011 war, Libya was one of the strongest economies within Africa. We found that some of the data points given were reported in 2008, where there was a global recession, which may have led to ambiguous information about the real state of these countries.

The uniqueness of our region, in comparison to the world, is not based on one issue but an entire environment. To understand the uniqueness of the North Sub-Saharan region, we chose variables from the 3 previously established sectors.
 
To compare the economy of our region to the world, we focused on the “Energy use (kg of oil equivalent) per $1000 GDP”.  This feature represents the annual energy usage in comparison to GDP, which is an indicator of a country’s development. Our region’s numbers rank significantly lower than the world’s average in this department, as expected.
 
On the Social side, the “Literacy Rate, adult total (% of people ages 15 and above)” showed that 3 of the 10 countries in our region have very low values. Mali, Chad, and Niger reported 26%, 28%, 29%, respectively. To compare, the average of the other regions from the world is 83.28%. Also, only 11 out of 100 people have access to the internet while the global average is slightly below 30%.
 
For the sector of Health, our team looked at the Prevalence of HIV and examined values ranging from 0.1% (which is also the mode) to 1.7%. After calculating the average for all other regions, we realized that in that section, our region was less impacted than others. The calculated average came out to 2.17% for the rest of the world. The second Health factor analyzed was the Prevalence of undernourishment (% of population). In that sector as well, we found out that Northern Sub-Saharan Africa had a wide range of values.
